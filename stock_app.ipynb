```python
import pandas as pd

stock_path="Resources/Current_Portfolio_Positions.csv"

df=pd.read_csv(stock_path).dropna()
df

#column to calcute distribution of funds
col="Today's Gain/Loss Percent"
col2="Today's Gain/Loss Dollar"

new_df=df.loc[:,df.columns.intersection(["Symbol",col,col2])]


new_df[col]=pd.to_numeric(new_df.loc[:,col].str.rstrip('%'),errors='coerce')

new_df[col2]=pd.to_numeric(new_df[col2].str.replace('\((.*)\)', '-\\1').str.replace('$', ''),errors='coerce')



negative_loss_df=new_df.loc[new_df[col]<0]



funds_to_invest=abs(negative_loss_df[col2].sum())



total_loss_percent=negative_loss_df.loc[:,col].sum()


negative_loss_df['%_of_sum_of_percent_loss']=negative_loss_df.loc[:,col]/total_loss_percent


negative_loss_df['funds_to_distribute']=round(negative_loss_df.loc[:,'%_of_sum_of_percent_loss']*funds_to_invest,2)

#negative_loss_df

#negative_loss_df["adjusted_funds_to_distribute"]=negative_loss_df["%_of_sum_of_percent_loss"]*round(negative_loss_df['funds_to_distribute'].sum()+len(negative_loss_df.loc[negative_loss_df["funds_to_distribute"]<1]),0)

negative_loss_df.loc[negative_loss_df["funds_to_distribute"]<1,"funds_to_distribute"]=1

negative_loss_df
```
